## Count Number of Digits in Integer

```python
# number = 5864
def countDigits(number):
    count = 0
    while number > 0:
        count += 1
        number = number // 10    
    return count

countDigits(48647)


# TC -->  O(log10(n))
# SC -->  O(1)
```

```python
# Using Logarithmic Code

import math

def countDigit(n):
    result = math.floor(math.log10(n)+1)
    return result

countDigit(5478786425)
 
```

## Armstrong Number


```python
def armstrong_no(n):
    num = n
    total = 0
    digit_n = len(str(n))
    while num > 0:
        last_digit = num % 10
        total = total + (last_digit**digit_n)
        num = num // 10
        
    return total == n

print(armstrong_no(1634))
```

## Print All Factors of a Given Number

```python
# Brute Force
def prime_factors(num):
    result = []
    for i in range(1, num+1):
        if num % i == 0:
            result.append(i)
    return result

prime_factors(10)

# TC --> O(N)
# SC --> O(k)
```

```python
# Better Solution

def is_prime(num):
    result = []
    for i in range(1,(num//2)+1):
        if num % i == 0:
            result.append(i)
            
    result.append(num)    # Appending the last value
    return result


is_prime(20)
```

```python
# optimal solution
from math import *
def is_prime(num):
    result = []
    for i in range(1,int(sqrt(num))+1):
        if num % i == 0:
            result.append(i)
            
        if num // i != i:
            result.append(num//i)
            
    result.sort()        
    return result

is_prime(20)

# TC --> O(root(N) + O(NlogN))
# SC --> O(k)
```
